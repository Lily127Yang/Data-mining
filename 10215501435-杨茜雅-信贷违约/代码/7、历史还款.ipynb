#### 2.1  对训练数据新增his_loan_amount_all和avg_period

```python
import pandas as pd

# 加载数据
train_both_data = pd.read_csv('训练1/train_both_with_merchant_info_with_store_count.txt')
train_history_data = pd.read_csv('训练1/train_history_with_merchant_info.txt')
history_loan_data = pd.read_csv('历史还款数据/table_datamining_history_loan.txt')

# 打印合并前的数据大小
print("合并前的train_both数据大小（行，列）:", train_both_data.shape)
print(train_both_data.head(3))
print("合并前的train_history数据大小（行，列）:", train_history_data.shape)
print(train_history_data.head(3))

# 定义函数计算总和和平均值
def calculate_stats(df):
    # 计算每个merchant_id的lend_amount总和和lend_period平均值
    grouped = df.groupby('merchant_id').agg(
        his_loan_amount_all=('lend_amount', 'sum'),
        avg_period=('lend_period', 'mean')
    ).reset_index()
    return grouped

# 对history_loan_data进行分组和计算
stats = calculate_stats(history_loan_data)

# 合并计算结果到训练数据中
train_both_updated = pd.merge(train_both_data, stats, on='merchant_id', how='left')
train_history_updated = pd.merge(train_history_data, stats, on='merchant_id', how='left')

# 打印合并后的数据大小
print("合并后的train_both数据大小（行，列）:", train_both_updated.shape)
print("合并后的train_history数据大小（行，列）:", train_history_updated.shape)

# 保存更新后的数据到新文件
train_both_updated.to_csv('训练1/train_both_his1.txt', index=False)
train_history_updated.to_csv('训练1/train_history_1', index=False)

# 打印合并后的数据前三行以确认
print("Updated train_both data:")
print(train_both_updated.head(3))
print("\nUpdated train_history data:")
print(train_history_updated.head(3))
```

### 3、新增overdue_count、loan_count、overdue_rate、overdue_days_all和overdue_sum五个特征。（历史违约次数，总贷款申请次数、历史违约率、历史违约付款的总天数，从理论和实际还款日推算出的违约次数）

#### 3.1  对训练数据新增overdue_count、loan_count、overdue_rate、overdue_days_all和overdue_sum五个特征

###### 删除apply_id,apply_date,lend_amount,lend_period,is_30days_overdue字段，只剩下merchant_id，notional_settle_date,real_settle_date字段。新增字段overdue、overdue_days，overdue_days_all和overdue_sum,对于每行数据，如果real_settle_date的日期晚于notional_settle_date的日期，说明逾期还款了，因此overdue字段设为1，real_settle_date和notional_settle_date之间间隔的天数就是overdue_days；如果real_settle_date的日期早于notional_settle_date的日期，说明没有逾期，因此overdue字段设为0，overdue_days也为0.对于字段overdue_days_all和overdue_sum，计算逻辑如下：如果该行的merchant_id没有在文件中重复过，则overdue_days_all字段的值就是overdue_days字段的值，overdue_sum字段的值就是overdue字段的值；如果该merchant_id有重复，则overdue_days_all字段的值是merchant_id相同的行的overdue_days字段求和，overdue_sum字段的值就是merchant_id相同的行的overdue字段求和。

```python
import pandas as pd

# 加载数据
train_both_data = pd.read_csv('训练1/train_both_his1.txt')
train_history_data = pd.read_csv('训练1/train_history_1.txt')
history_loan_data = pd.read_csv('历史还款数据/table_datamining_history_loan.txt')

print("合并前的train_both数据大小（行，列）:", train_both_data.shape)
print(train_both_data.head(3))
print("合并前的train_history数据大小（行，列）:", train_history_data.shape)
print(train_history_data.head(3))

# 确保所有数据框中merchant_id不作为索引
train_both_data.reset_index(drop=True, inplace=True)
train_history_data.reset_index(drop=True, inplace=True)
history_loan_data.reset_index(drop=True, inplace=True)

# 定义函数计算总和和平均值
def calculate_loan_stats(df):
    grouped = df.groupby('merchant_id').agg(
        loan_count=('apply_id', 'nunique'),
        overdue_count=('is_30days_overdue', lambda x: (x == 1).sum())
    ).reset_index()
    grouped['overdue_rate'] = grouped['overdue_count'] / grouped['loan_count']
    return grouped

# 定义计算逾期统计数据的函数
def calculate_overdue_stats(df):
    df['overdue'] = (pd.to_datetime(df['real_settle_date']) > pd.to_datetime(df['notional_settle_date'])).astype(int)
    df['overdue_days'] = (pd.to_datetime(df['real_settle_date']) - pd.to_datetime(df['notional_settle_date'])).dt.days.clip(lower=0)
    grouped = df.groupby('merchant_id').agg(
        overdue_days_all=('overdue_days', 'sum'),
        overdue_sum=('overdue', 'sum')
    ).reset_index()
    return grouped

# 计算统计数据
loan_stats = calculate_loan_stats(history_loan_data)
overdue_stats = calculate_overdue_stats(history_loan_data)

# 合并计算结果到训练数据中
train_both_updated = pd.merge(train_both_data, loan_stats, on='merchant_id', how='left')
train_both_updated = pd.merge(train_both_updated, overdue_stats, on='merchant_id', how='left')
train_history_updated = pd.merge(train_history_data, loan_stats, on='merchant_id', how='left')
train_history_updated = pd.merge(train_history_updated, overdue_stats, on='merchant_id', how='left')

# 保存更新后的数据到新文件
train_both_updated.to_csv('训练1/train_both_his2.txt', index=False)
train_history_updated.to_csv('训练1/train_history_2.txt', index=False)

# 打印合并后的数据规模和前三行
print("合并后的train_both数据大小（行，列）:", train_both_updated.shape)
print(train_both_updated.head(3))
print("合并后的train_history数据大小（行，列）:", train_history_updated.shape)
print(train_history_updated.head(3))
```

```python
import pandas as pd

# 定义函数用于处理和比较数据
def process_and_compare(train_data_path, history_data_path):
    # 加载数据
    train_data = pd.read_csv(train_data_path)
    history_data = pd.read_csv(history_data_path)

    # 提取history_data中的merchant_id及对应统计字段
    required_fields = ['merchant_id', 'his_loan_amount_all', 'avg_period', 'loan_count', 'overdue_count', 'overdue_rate', 'overdue_days_all', 'overdue_sum']
    history_stats = history_data[required_fields]

    # 合并数据以比较字段值
    comparison_df = pd.merge(train_data, history_stats, on='merchant_id', how='inner', suffixes=('_train', '_history'))

    # 检查所有统计字段是否相同
    fields_to_check = ['his_loan_amount_all', 'avg_period', 'loan_count', 'overdue_count', 'overdue_rate', 'overdue_days_all', 'overdue_sum']
    mismatches = {field: comparison_df[comparison_df[f"{field}_train"] != comparison_df[f"{field}_history"]] for field in fields_to_check}

    # 输出结果
    print(f"处理文件：{train_data_path}")
    for field, mismatches_df in mismatches.items():
        if not mismatches_df.empty:
            print(f"字段 {field} 在一些记录中不匹配。不匹配数量: {len(mismatches_df)}")
        else:
            print(f"字段 {field} 在所有记录中都匹配。")

    # 打印合并后数据的总体规模和前三行，以确保正确加载和合并
    print("合并后数据的规模（行数，列数）:", comparison_df.shape)
    print("合并后的数据前三行:")
    print(comparison_df.head(3))
    print("\n")  # 添加空行以分隔不同文件的输出

# 定义数据路径
train_both_path = '测试1/test_both_his2.txt'
train_history_path = '测试1/test_history_2.txt'
history_data_path = '历史还款数据/history_2.txt'

# 处理两个文件
process_and_compare(train_both_path, history_data_path)
process_and_compare(train_history_path, history_data_path)
```

```python
import pandas as pd

# 加载数据
history_data_path = '历史还款数据/history_2.txt'
train_history_path = '训练1/train_history_2.txt'

history_data = pd.read_csv(history_data_path)
train_history = pd.read_csv(train_history_path)

# 找出两个文件中相同的merchant_id
common_merchant_ids = set(history_data['merchant_id']).intersection(set(train_history['merchant_id']))
print(f"两个文件中相同的Merchant_id数量: {len(common_merchant_ids)}")

# 删除history_data中的这些merchant_id
original_history_size = len(history_data)
history_data = history_data[~history_data['merchant_id'].isin(common_merchant_ids)]
print(f"删除指定merchant_id后，history_data中剩余的行数: {len(history_data)} (原始行数: {original_history_size})")

# 合并剩余的数据与train_history
combined_data = pd.concat([train_history, history_data], ignore_index=True)

# 保存合并后的数据到新文件
output_path = '历史还款数据/history_3.txt'
combined_data.to_csv(output_path, index=False)

# 打印合并前后的数据规模
print(f"合并前train_history的数据规模: {train_history.shape}")
print(f"合并后的数据规模（history_3.txt）: {combined_data.shape}")
```
