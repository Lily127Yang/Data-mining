### 2、文件分割依据

#### 2.1 训练数据

```python
import pandas as pd

# 加载数据
df_loan_train = pd.read_csv('贷款数据/table_loan_train.txt')
df_combined_data = pd.read_csv('流水数据/clean1.txt')
df_history_loan = pd.read_csv('历史还款数据/table_datamining_history_loan.txt')

# 提取各数据集中的merchant_id
in_combined = set(df_combined_data['merchant_id'].unique())
in_history = set(df_history_loan['merchant_id'].unique())

# 计算各种情况
count_1 = count_2 = count_3 = count_4 = 0

for merchant_id in df_loan_train['merchant_id']:
    if merchant_id in in_combined and merchant_id in in_history:
        count_1 += 1
    elif merchant_id in in_combined and merchant_id not in in_history:
        count_2 += 1
    elif merchant_id not in in_combined and merchant_id in in_history:
        count_3 += 1
    elif merchant_id not in in_combined and merchant_id not in in_history:
        count_4 += 1

# 输出结果
print("对于训练数据而言:")
print("在两个数据集中都找到的merchant_id数量:", count_1)
print("只在流水数据中找到的merchant_id数量:", count_2)
print("只在历史还款数据中找到的merchant_id数量:", count_3)
print("在两个数据集中都找不到的merchant_id数量:", count_4)

# 输出四个规则的总merchant_id数量
total_merchant_ids = count_1 + count_2 + count_3 + count_4
print("四个规则的总merchant_id数量:", total_merchant_ids)
```

```python
import pandas as pd

# 加载数据
df_loan_train = pd.read_csv('贷款数据/table_loan_train.txt')
df_combined_data = pd.read_csv('流水数据/clean1.txt')
df_history_loan = pd.read_csv('历史还款数据/table_datamining_history_loan.txt')

# 提取各数据集中的merchant_id
in_combined = set(df_combined_data['merchant_id'].unique())
in_history = set(df_history_loan['merchant_id'].unique())

# 创建四个空的DataFrame用于存储各类数据
df_1 = pd.DataFrame(columns=df_loan_train.columns)
df_2 = pd.DataFrame(columns=df_loan_train.columns)
df_3 = pd.DataFrame(columns=df_loan_train.columns)
df_4 = pd.DataFrame(columns=df_loan_train.columns)

# 计算各种情况并分配数据
for index, row in df_loan_train.iterrows():
    merchant_id = row['merchant_id']
    if merchant_id in in_combined and merchant_id in in_history:
        df_1 = df_1.append(row)
    elif merchant_id in in_combined and merchant_id not in in_history:
        df_2 = df_2.append(row)
    elif merchant_id not in in_combined and merchant_id in in_history:
        df_3 = df_3.append(row)
    elif merchant_id not in in_combined and merchant_id not in in_history:
        df_4 = df_4.append(row)

# 保存数据到不同的文件
df_1.to_csv('训练1/merchant_in_both_datasets.txt', index=False)
df_2.to_csv('训练1/merchant_only_in_combined.txt', index=False)
df_3.to_csv('训练1/merchant_only_in_history.txt', index=False)
df_4.to_csv('训练1/merchant_in_neither_dataset.txt', index=False)

# 输出结果
print("在两个数据集中都找到的merchant_id数量:", len(df_1))
print("只在流水数据中找到的merchant_id数量:", len(df_2))
print("只在历史还款数据中找到的merchant_id数量:", len(df_3))
print("在两个数据集中都找不到的merchant_id数量:", len(df_4))

# 输出四个规则的总merchant_id数量
total_merchant_ids = len(df_1) + len(df_2) + len(df_3) + len(df_4)
print("四个规则的总merchant_id数量:", total_merchant_ids)
```

#### 2.2 训练数据分割check

```python
import pandas as pd

# 加载四个分类的数据文件
df_both = pd.read_csv('训练1/merchant_in_both_datasets.txt')
df_combined = pd.read_csv('训练1/merchant_only_in_combined.txt')
df_history = pd.read_csv('训练1/merchant_only_in_history.txt')
df_neither = pd.read_csv('训练1/merchant_in_neither_dataset.txt')

# 合并这四个DataFrame
df_merged = pd.concat([df_both, df_combined, df_history, df_neither], ignore_index=True)

# 加载原始的训练数据
df_original = pd.read_csv('训练1/table_loan_train.txt')

# 对两个DataFrame的行进行排序，并重置索引，以便比较
df_merged_sorted = df_merged.sort_values(by=list(df_merged.columns)).reset_index(drop=True)
df_original_sorted = df_original.sort_values(by=list(df_original.columns)).reset_index(drop=True)

# 比较两个DataFrame
are_identical = df_merged_sorted.equals(df_original_sorted)

# 输出比较结果
print("合并后的文件内容与原始贷款数据文件内容是否完全一样:", are_identical)
```

#### 2.5 预测文件

```python
import pandas as pd

# 加载数据
df_predict = pd.read_csv('预测1/table_loan_predict.txt')
df_test_both = pd.read_csv('测试1/test_merchant_in_both_datasets.txt')
df_test_combined = pd.read_csv('测试1/test_merchant_only_in_combined.txt')
df_test_history = pd.read_csv('测试1/test_merchant_only_in_history.txt')
df_test_neither = pd.read_csv('测试1/test_merchant_in_neither_dataset.txt')

# 使用 apply_id 进行合并，创建对应的预测数据文件
df_predict_both = pd.merge(df_predict, df_test_both['apply_id'], on='apply_id', how='inner')
df_predict_combined = pd.merge(df_predict, df_test_combined['apply_id'], on='apply_id', how='inner')
df_predict_history = pd.merge(df_predict, df_test_history['apply_id'], on='apply_id', how='inner')
df_predict_neither = pd.merge(df_predict, df_test_neither['apply_id'], on='apply_id', how='inner')

# 保存到文件
df_predict_both.to_csv('预测1/predict_merchant_in_both_datasets.txt', index=False)
df_predict_combined.to_csv('预测1/predict_merchant_only_in_combined.txt', index=False)
df_predict_history.to_csv('预测1/predict_merchant_only_in_history.txt', index=False)
df_predict_neither.to_csv('预测1/predict_merchant_in_neither_dataset.txt', index=False)

print("预测数据已按对应的分类保存。")
```
